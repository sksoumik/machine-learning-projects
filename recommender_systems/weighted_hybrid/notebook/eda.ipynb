```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('../data/tmdb_5000_movies.csv')
credits = pd.read_csv('../data/tmdb_5000_credits.csv')
```

```python
movies.head(2)
```

```python
credits.head(2)
```

```python
# rename columns
credit_column_renamed = credits.rename(index=str, columns={'movie_id': 'id'})
movies_df_merged = movies.merge(credit_column_renamed, on='id')
```

```python
movies_df_merged.head(2)
```

```python
movies_cleaned_df = movies_df_merged.drop(columns=['homepage', 'title_x', 'title_y', 'status','production_countries'])
movies_cleaned_df.head(2)
```

```python
movies_cleaned_df.info()
```

```python
# calculate weighted average of movies based 
v = movies_cleaned_df['vote_count']
r = movies_cleaned_df['vote_average']
c = movies_cleaned_df['vote_average'].mean()
m = movies_cleaned_df['vote_count'].quantile(0.70)

movies_cleaned_df['weighted_average'] = ((r*v)+(c*m))/(v+m)
```

```python
movies_cleaned_df.head()
```

```python
movie_sorted_ranking = movies_cleaned_df.sort_values('weighted_average', ascending=False)
movie_sorted_ranking[['original_title', 'vote_count', 'vote_average', 'weighted_average', 'popularity']].head(10)
```
