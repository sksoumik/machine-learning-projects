```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

### Read CSV file using Panda Dataframe

```python
df = pd.read_csv("movie_dataset.csv")
```

```python
df.head()
```

```python
print("Total samples", df.shape)
print("column names: \n", df.columns)
```

We don't need all of them, we just need the columns: keywords, cast, genres, director

```python
features = ['keywords','cast','genres','director']
```

```python
# Combine all the values from the above selected columns in a single string
def combine_features(row):
    return row['keywords']+" "+row['cast']+" "+row['genres']+" "+row['director']
```

### Pre-processing

```python
# fill all NaN values with blank string
for i in features:
    df[i] = df[i].fillna('')
```

```python
df["combined_features"] = df.apply(combine_features,axis=1)
```

```python
df.iloc[0].combined_features
```

```python
df.iloc[2].combined_features
```

### Count Matrix

```python
# feed these strings to a CountVectorizer() object for getting the count matrix.
cv = CountVectorizer()
count_matrix = cv.fit_transform(df["combined_features"])
```

### Find cosine similarity

```python
cosine_sim = cosine_similarity(count_matrix)
```

### Get movie title from movie index 

```python
def get_title_from_index(index):
    return df[df.index == index]["title"].values[0]
```

###  Get movie index from movie title

```python
def get_index_from_title(title):
    return df[df.title == title]["index"].values[0]
```

```python
movie_user_likes = "The Dark Knight Rises"
movie_index = get_index_from_title(movie_user_likes)
```

```python
movie_index
```

```python
similar_movies = list(enumerate(cosine_sim[movie_index]))
```

```python
similar_movies
```

### Sort this list in descending order

```python
sorted_similar_movies = sorted(similar_movies,key=lambda x:x[1],reverse=True)[1:]
```

```python
sorted_similar_movies
```

### Top 10 out of them

```python
i=0
print("Top 10 similar movies to "+movie_user_likes+" are:\n")
for element in sorted_similar_movies:
    print(get_title_from_index(element[0]))
    i=i+1
    if i>10:
        break
```
